```python
import FileWorker as FW
import Visualisation as vis
import Gabor as gb
```

```python
path = "fp1.png"
fw = FW.FileWorker()
img = fw.ReadImg(path)
angles = [i for i in range(0,181)]
```

```python
title="K_size=10, Sigma=8, Gamma=1.5, Lambda=6"
out = gb.Gabor_process(img,K_size=10, Sigma=8, Gamma=1.5, Lambda=6,angles=angles)
vis.CompareImages(img,out,True,'Results/',names=['Input','Result1'],title=title,save_as=1)
```
